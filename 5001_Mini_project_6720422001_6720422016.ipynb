```python
import pandas as pd
import plotly.express as px

df_chipotle = pd.read_csv(
    'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv',
    sep='\t'
)

# แปลงราคาให้เป็นตัวเลข
df_chipotle['item_price'] = df_chipotle['item_price'].str.replace('$','').astype(float)

# สร้าง revenue ต่อรายการ
df_chipotle['revenue'] = df_chipotle['quantity'] * df_chipotle['item_price']
```

```python
# หัวข้อ 1 (Average order value)

order_value = df_chipotle.groupby('order_id', as_index=False)['revenue'].sum()

fig = px.histogram(order_value, x='revenue', nbins=100,
                   title='Distribution of Order Values (Chipotle)',
                   labels={'revenue':'Order Value ($)'})
fig.show()
```

```python
# หัวข้อ 2 (เมนูยอดนิยม & รายได้)

menu_summary = df_chipotle.groupby('item_name', as_index=False)['quantity'].sum()
top_menu = menu_summary.sort_values('quantity', ascending=False).head(10)

fig = px.bar(top_menu,
             x='quantity',
             y='item_name',
             orientation='h',
             title='Top 10 Most Ordered Menu Items',
             labels={'quantity':'Total Quantity','item_name':'Menu Item'},
             color='quantity',
             color_continuous_scale='teal')
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```

```python
# หัวข้อ 2 (เมนูยอดนิยม & รายได้)

revenue_summary = df_chipotle.groupby('item_name', as_index=False)['revenue'].sum()
top_revenue = revenue_summary.sort_values('revenue', ascending=False).head(10)

fig = px.bar(top_revenue,
             x='revenue', y='item_name',
             orientation='h',
             title='Top 10 Menu Items by Revenue',
             color='revenue', color_continuous_scale='sunset')
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```

```python
import pandas as pd
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# โหลดข้อมูล
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
df['item_price'] = df['item_price'].str.replace('$','').astype(float)

# คำนวณยอดขายต่อรายการ
df['total_price'] = df['quantity'] * df['item_price']

# 1️⃣ เมนูยอดนิยม (ตามจำนวนครั้งที่ถูกสั่ง)
top_items = (
    df.groupby('item_name')['quantity']
    .sum()
    .reset_index()
    .sort_values(by='quantity', ascending=False)
    .head(10)
)

# 2️⃣ เมนูที่ทำรายได้สูงสุด
top_revenue = (
    df.groupby('item_name')['total_price']
    .sum()
    .reset_index()
    .sort_values(by='total_price', ascending=False)
    .head(10)
)

# 3️⃣ ทำ Dashboard รวม 2 กราฟ
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=('Top 10 Most Ordered Items', 'Top 10 Highest Revenue Items')
)

# กราฟซ้าย: ยอดสั่ง
fig.add_trace(
    go.Bar(
        x=top_items['quantity'],
        y=top_items['item_name'],
        orientation='h',
        marker=dict(color=top_items['quantity'], colorscale='mint'),
        name='Most Ordered'
    ),
    row=1, col=1
)

# กราฟขวา: รายได้
fig.add_trace(
    go.Bar(
        x=top_revenue['total_price'],
        y=top_revenue['item_name'],
        orientation='h',
        marker=dict(color=top_revenue['total_price'], colorscale='teal'),
        name='Top Revenue'
    ),
    row=1, col=2
)

# Layout
fig.update_layout(
    title_text='Chipotle Menu Analysis Dashboard',
    height=600,
    showlegend=False,
    plot_bgcolor='white',
    hovermode='y unified'
)

fig.update_yaxes(categoryorder='total ascending', row=1, col=1)
fig.update_yaxes(categoryorder='total ascending', row=1, col=2)

fig.show()
```

```python
# หัวข้อ 3 (Bundle/co-occurrence)

import pandas as pd
import plotly.express as px
import itertools

# โหลดข้อมูล
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
df['item_price'] = df['item_price'].str.replace('$','').astype(float)

# กลุ่มตาม order_id → รายการอาหารในแต่ละคำสั่งซื้อ
order_items = df.groupby('order_id')['item_name'].apply(list).reset_index()
order_items.head()
```

```python
from collections import Counter

pair_counts = Counter()

for items in order_items['item_name']:
    unique_items = list(set(items))
    pairs = itertools.combinations(unique_items, 2)
    pair_counts.update(pairs)

df_pairs = pd.DataFrame(pair_counts.items(), columns=['pair', 'count'])
df_pairs[['item1', 'item2']] = pd.DataFrame(df_pairs['pair'].tolist(), index=df_pairs.index)
df_pairs = df_pairs.drop(columns='pair').sort_values('count', ascending=False)
df_pairs.head(10)
```

```python
top_pairs = df_pairs.head(10)

fig = px.bar(top_pairs,
             x='count',
             y=top_pairs.apply(lambda x: f"{x['item1']} + {x['item2']}", axis=1),
             orientation='h',
             title='Top 10 Menu Pairs Frequently Ordered Together',
             labels={'count':'Frequency', 'y':'Menu Pair'},
             color='count',
             color_continuous_scale='mint')
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```

```python
import networkx as nx
import plotly.graph_objects as go

# ใช้เฉพาะคู่ที่สั่งเกิน threshold
threshold = 15
filtered_pairs = df_pairs[df_pairs['count'] >= threshold]

# สร้าง graph
G = nx.Graph()
for _, row in filtered_pairs.iterrows():
    G.add_edge(row['item1'], row['item2'], weight=row['count'])

pos = nx.spring_layout(G, k=0.6, seed=42)

# เตรียมข้อมูลสำหรับ plotly
edge_x, edge_y = [], []
for edge in G.edges():
    x0, y0 = pos[edge[0]]
    x1, y1 = pos[edge[1]]
    edge_x += [x0, x1, None]
    edge_y += [y0, y1, None]

edge_trace = go.Scatter(
    x=edge_x, y=edge_y,
    line=dict(width=0.5, color='#888'),
    hoverinfo='none', mode='lines')

node_x, node_y, text, size = [], [], [], []
for node in G.nodes():
    x, y = pos[node]
    node_x.append(x)
    node_y.append(y)
    text.append(node)
    size.append(G.degree[node]*2)

node_trace = go.Scatter(
    x=node_x, y=node_y, mode='markers+text',
    text=text, textposition='top center',
    hoverinfo='text',
    marker=dict(size=size, color=size, colorscale='teal', line_width=1.5))

fig = go.Figure(data=[edge_trace, node_trace],
                layout=go.Layout(
                    title='Chipotle Menu Co-Occurrence Network',
                    showlegend=False,
                    hovermode='closest'))
fig.show()
```
